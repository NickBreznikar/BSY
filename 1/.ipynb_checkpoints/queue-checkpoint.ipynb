```python
import numpy as np
import math


def expr(input):
    return -math.log(np.random.uniform(0,1) / input)

serviceIntervals = []

amountOfArrivals = 100

for x in range(amountOfArrivals):
    serviceIntervals.append(expr(1))
    
arrivals = []
for x in range(amountOfArrivals):
    if x > 0:
        arrivals.append(expr(1) + arrivals[x - 1])
    else:
        arrivals.append(expr(1))
        
        
waitingTimes = []

serviceEnd = 0

for next in range(len(arrivals)):
    
    activeIntervalTime =  serviceIntervals[next]
    serviceStart = max([serviceEnd, arrivals[next]])
    waitingTime = max([0, serviceEnd - arrivals[next]])
    waitingTimes.append(waitingTime)
    serviceEnd = serviceStart + activeIntervalTime
    
    print("Ankunft #" + str(next) + " um: " + str(arrivals[next]) + " nächste Bearbeitungszeit: " + str(activeIntervalTime))
    print("Abreise #" + str(next) + " um: " + str(serviceEnd) + " Wartezeit: " + str(waitingTime))

sum = np.sum(waitingTimes)
#expextedValue = 0,5
middle = sum/amountOfArrivals

print("")
    
print("Für " + str(amountOfArrivals) + " Ankünfte:")
print("Mittelwert: " + str(middle))
print("Varianz: " + str(np.var(waitingTimes)))
```
